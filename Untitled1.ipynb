```python
import pandas as pd
import numpy as np
```

```python
from collections import Counter
```

```python
raw_data = pd.read_csv('data/international_matches_1950_2017.csv')
```

```python
raw_data = raw_data.sort_values(['date'])
```

```python
set(raw_data['CupName'])
```

```python
match_types_keep = [ 'Continental Final',
                     'Continental Qualifier',
                     'FIFA Confederations Cup',
                     'FIFA World Cup™ Final',
                     'FIFA World Cup™ Qualifier']
```

```python
matches_no_friendlys = raw_data.loc[raw_data['CupName'].isin(match_types_keep)]
```

```python
matches_no_friendlys['goal_difference'] = matches_no_friendlys.apply(lambda row: row['team1Score'] - row['team2Score'], axis=1)
```

```python
def winner(row):
    if row['goal_difference'] < 0:
        return -1
    elif row['goal_difference'] == 0:
        return 0
    elif row['goal_difference'] > 0:
        return 1
    else:
        return np.NaN
    
matches_no_friendlys['winner'] = matches_no_friendlys.apply(winner, axis=1)
```

```python
matches_no_friendlys.describe()
```

```python
matches_exclude_outliers = matches_no_friendlys.loc[np.abs(matches_no_friendlys['goal_difference']) < 7]
```

```python
cols = list(matches_exclude_outliers.columns)
matches_exclude_outliers_reverse = matches_exclude_outliers.copy()

matches_exclude_outliers_reverse['team1'] = matches_exclude_outliers['team2']
matches_exclude_outliers_reverse['team2'] = matches_exclude_outliers['team1']
matches_exclude_outliers_reverse['team1Text'] = matches_exclude_outliers['team2Text']
matches_exclude_outliers_reverse['team2Text'] = matches_exclude_outliers['team1Text']

matches_exclude_outliers_reverse['team1Score'] = matches_exclude_outliers['team2Score']
matches_exclude_outliers_reverse['team2Score'] = matches_exclude_outliers['team1Score']
matches_exclude_outliers_reverse['team1PenScore'] = matches_exclude_outliers['team2PenScore']
matches_exclude_outliers_reverse['team2PenScore'] = matches_exclude_outliers['team1PenScore']

matches_exclude_outliers_reverse['goal_difference'] = matches_exclude_outliers['goal_difference'] * -1
matches_exclude_outliers_reverse['winner'] = matches_exclude_outliers['winner'] * -1

matches_exclude_outliers_reverse['resText'] = matches_exclude_outliers['resText']
```

```python
def reverseScore(row):
#     if '(' in score:
#         return '{} ({})'.format(score.split(' ')[0], score.split('(')[1][:-1][::-1])
#     return score[::-1]
    if row['statText'] == 'Win on penalty':
        return '{}-{} ({}-{})'.format(row['team1Score'], row['team2Score'], int(row['team1PenScore']), int(row['team2PenScore']))
    return '{}-{}'.format(row['team1Score'], row['team2Score'])
    
# matches_exclude_outliers['resText'] = matches_exclude_outliers.apply(reverseScore)
```

```python
matches_exclude_outliers_reverse['resText'] = matches_exclude_outliers_reverse.apply(reverseScore, axis=1)
```
