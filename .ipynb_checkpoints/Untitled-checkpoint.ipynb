```python
match_features = pd.read_csv('data/match_features.csv')
#convert date to timestamp
match_features['date'] = pd.to_datetime(match_features['date'])

match_features = match_features.sort_values('date')

# match_features['winner'] = match_features.apply(lambda row: 0 if row['outcome'] == 0 else (row['outcome'] / np.abs(row['outcome'])), axis=1)

# train_data, test_data = match_features.head(15000), match_features.tail(match_features.shape[0]-15000) 

X_train, X_test, y_train, y_test = train_test_split(match_features, match_features.outcome, test_size=0.1)
# X_train, X_test, y_train, y_test = train_test_split(match_features, match_features.winner, test_size=0.1)

X_train = X_train.drop(['outcome', 'date', 'team1', 'team2', 'team1Home', 'team2Home', 'neutralVenue', 'friendly', 'qualifier', 'finaltourn'], axis=1)
X_test = X_test.drop(['outcome', 'date', 'team1', 'team2', 'team1Home', 'team2Home', 'neutralVenue', 'friendly', 'qualifier', 'finaltourn'], axis=1)
# X_train = X_train.drop(['winner', 'date', 'team1', 'team2', 'team1Home', 'team2Home', 'neutralVenue', 'friendly', 'qualifier', 'finaltourn'], axis=1)
# X_test = X_test.drop(['winner', 'date', 'team1', 'team2', 'team1Home', 'team2Home', 'neutralVenue', 'friendly', 'qualifier', 'finaltourn'], axis=1)
```

```python
model = Sequential()
model.add(Dense(256, input_dim=X_train.shape[1], activation='sigmoid'))
model.add(Dropout(0.25))
model.add(Dense(512, activation='sigmoid'))
model.add(Dropout(0.25))
model.add(Dense(3, activation='sigmoid'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(X_train, y_train, epochs=32, batch_size=64)
```

```python
score = model.evaluate(X_test, y_test, batch_size=16)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, n_jobs=-1, verbose=1)
```

```python
# Train the model on training data
rf.fit(X_train, y_train)
```

```python
predictions = []
for index, values in matchups_topredict.iterrows():
    match = []
    for tree in rf.estimators_:
        pred = tree.predict([values])
        match.append(pred[0])
    predictions.append(match)
```

```python
power_ranking = {
    'RUS': 1/32,
    'KSA': 1/256,
    'EGY': 1/100,
    'URU': 1/28,
    'POR': 1/22,
    'ESP': 1/6,
    'MAR': 1/256,
    'IRN': 1/256,
    'FRA': 1/6,
    'AUS': 1/150,
    'PER': 1/150,
    'DEN': 1/45,
    'ARG': 1/9,
    'ISL': 1/100,
    'CRO': 1/25,
    'NGA': 1/256,
    'BRA': 1/4,
    'SUI': 1/66,
    'CRC': 1/256,
    'SRB': 1/256,
    'GER': 1/5,
    'SWE': 1/125,
    'MEX': 1/66,
    'KOR': 1/256,
    'BEL': 1/11,
    'PAN': 1/256,
    'TUN': 1/256,
    'ENG': 1/16,
    'POL': 1/35,
    'SEN': 1/256,
    'COL': 1/33,
    'JPN': 1/128 
}
```

```python
with open('predicted_matchups_static_with_power_ranking.csv', 'w') as out_file:
    for i, matchup in enumerate(predictions):
        mean = np.mean(matchup)
        std = np.std(matchup)
        t1 = matchups_topredict_raw.iloc[i]['team1']
        t2 = matchups_topredict_raw.iloc[i]['team2']
        
        mean += power_ranking[t1] - power_ranking[t2]
        std  -= np.abs(power_ranking[t1] - power_ranking[t2])
                
        out_file.write('{},{},{},{}\n'.format(t1, t2, mean, std))
```
